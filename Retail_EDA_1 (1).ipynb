```python
import pandas as pd
import io

df = pd.read_csv("retail_sales_messy_100K.csv", 
                  encoding='latin1', 
                  on_bad_lines='skip', 
                  engine='python',
                  encoding_errors='ignore')
```

## Step 4 — Check Duplicate Rows

## Step 5 — Check Data Types

```python
print(df.dtypes)

```

## Step 6 — Check Negative & Zero Values

```python
print('=== INVALID NUMERIC VALUES ===')

# Negative quantities (data entry errors)
neg_qty = df[df['Quantity'] < 0]
print(f'Negative Quantity rows: {len(neg_qty):,}')

# Zero unit price (invalid records)
zero_price = df[df['UnitPrice'] == 0]
print(f'Zero UnitPrice rows: {len(zero_price):,}')

# Discount > 1 (entered as 20 instead of 0.20)
bad_discount = df[df['Discount'] > 1]
print(f'Bad Discount (>1) rows: {len(bad_discount):,}')

# Negative profit (should investigate)
neg_profit = df[df['Profit'] < 0]
print(f'Negative Profit rows: {len(neg_profit):,}')

# Show sample bad rows
print('\n--- Sample rows with Negative Quantity ---')
df[df['Quantity'] < 0][['OrderID','OrderDate','ProductName','Quantity','UnitPrice','Revenue']].head()
```

## Step 7 — Check Unique Values in Categorical Columns

```python
# These columns should have fixed/limited unique values
# If you see typos or extra values — that's a problem

cat_cols = ['Region', 'Category', 'SubCategory', 'Segment', 'Channel', 'PaymentMethod']

for col in cat_cols:
    unique_vals = df[col].dropna().unique()
    print(f'\n{col} ({len(unique_vals)} unique values):')
    print(sorted(unique_vals))
```

## Step 8 — Spot Region Typos Specifically

```python
# Count each unique Region value
# Clean data should have only 5: North South East West Central
print('=== REGION VALUE COUNTS ===')
print(df['Region'].value_counts(dropna=False))

# How many rows have Region typos (not in the correct 5 values)
valid_regions = ['North', 'South', 'East', 'West', 'Central']
typo_rows = df[~df['Region'].isin(valid_regions) & df['Region'].notna()]
print(f'\nRows with Region typos: {len(typo_rows):,}')
print('Typo values found:')
print(typo_rows['Region'].value_counts())
```

## Step 9 — Check Date Column Issues

```python
# Try to convert OrderDate to datetime
# Rows that FAIL to convert are the inconsistent date formats
date_converted = pd.to_datetime(df['OrderDate'], errors='coerce')

# Count rows where conversion failed (NaT = Not a Time)
bad_dates = date_converted.isna().sum()
print(f'Rows with unparseable date format: {bad_dates:,}')

# Show sample bad date values
bad_date_rows = df[date_converted.isna()]
if len(bad_date_rows) > 0:
    print('\nSample bad date values:')
    print(bad_date_rows['OrderDate'].head(10).tolist())
else:
    print('All dates parsed successfully!')
```

## Step 10 — Check Extra Whitespace in ProductName

```python
# Check for leading/trailing spaces in ProductName
# A product with spaces looks identical but is treated as different

has_whitespace = df['ProductName'].dropna().apply(
    lambda x: x != x.strip()
).sum()

print(f'ProductName rows with extra whitespace: {has_whitespace:,}')

# Show examples
ws_examples = df[df['ProductName'].notna() & 
                 (df['ProductName'] != df['ProductName'].str.strip())]
print('\nSample whitespace values (shown with quotes):')
for val in ws_examples['ProductName'].head(5).tolist():
    print(f'  --> "{val}"')
```

## Step 11 — Statistical Summary

```python
# Summary statistics for numeric columns
# Look for: min values that are negative, max values that look wrong
df[['Quantity','UnitPrice','Discount','Revenue','COGS','Profit','ProfitMargin']].describe().round(2)
```

## Step 12 — Full EDA Summary Report

```python
print('=' * 55)
print('       FULL EDA SUMMARY REPORT')
print('=' * 55)

print(f'\n DATASET SIZE')
print(f'  Total Rows        : {len(df):,}')
print(f'  Total Columns     : {df.shape[1]}')

print(f'\n MISSING VALUES')
print(f'  CustomerID nulls  : {df["CustomerID"].isnull().sum():,}')
print(f'  ProductName nulls : {df["ProductName"].isnull().sum():,}')
print(f'  Region nulls      : {df["Region"].isnull().sum():,}')

print(f'\n DUPLICATE ROWS')
print(f'  Duplicate OrderIDs: {df.duplicated(subset=["OrderID"]).sum():,}')

print(f'\n INVALID NUMERIC VALUES')
print(f'  Negative Quantity : {(df["Quantity"] < 0).sum():,}')
print(f'  Zero UnitPrice    : {(df["UnitPrice"] == 0).sum():,}')
print(f'  Bad Discount (>1) : {(df["Discount"] > 1).sum():,}')

print(f'\n INCONSISTENT TEXT')
valid_regions = ['North', 'South', 'East', 'West', 'Central']
typos = df[~df['Region'].isin(valid_regions) & df['Region'].notna()]
print(f'  Region typos      : {len(typos):,}')
ws = df[df['ProductName'].notna() & (df['ProductName'] != df['ProductName'].str.strip())]
print(f'  Whitespace names  : {len(ws):,}')

print(f'\n DATE ISSUES')
bad_dt = pd.to_datetime(df['OrderDate'], errors='coerce').isna().sum()
print(f'  Bad date formats  : {bad_dt:,}')

total_issues = (
    df['CustomerID'].isnull().sum() +
    df['ProductName'].isnull().sum() +
    df['Region'].isnull().sum() +
    df.duplicated(subset=['OrderID']).sum() +
    (df['Quantity'] < 0).sum() +
    (df['UnitPrice'] == 0).sum() +
    (df['Discount'] > 1).sum() +
    len(typos) + len(ws) + bad_dt
)

print(f'\n TOTAL ISSUES FOUND : {total_issues:,}')
print(f' CLEAN ROWS (est.)  : {len(df) - total_issues:,}')
print('=' * 55)
print('  Next Step --> Clean all issues in Excel or Python')
print('=' * 55)
```
